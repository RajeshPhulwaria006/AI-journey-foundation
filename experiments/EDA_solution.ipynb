```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import kagglehub
path = kagglehub.dataset_download("brendan45774/test-file")
```

```python
df = pd.read_csv(path + "/tested.csv")
df.head(10)
```

1. Categorical dataset :

```python
# visualize frequency of survived peoples
print("Total passengers : ", df.shape[0])
print(df['Survived'].value_counts())
sns.countplot(x="Survived", data=df)
```

```python
# PieChart visualization
print(df['Sex'].value_counts().plot(kind='pie', autopct='%.2f'))
```

2. Numerical data.

```python
# Histogram
plt.hist(df['Age'])
```

```python
# Histogram + KDE [kernel/probability density estimation]
sns.distplot(df['Age'])
```

```python
# Boxplot - outlier identification
sns.boxplot(x="Age", data=df)
```

```python
# multiple boxplot
sns.boxplot(x="Pclass", y="Age", data=df)
```

> Bivariate analysis

```python
tipsDF = sns.load_dataset("tips")
flightDF = sns.load_dataset("flights")
irisDF = sns.load_dataset("iris")
titanicDF = pd.read_csv(path + "/tested.csv")
```

```python
# scatter plot
sns.scatterplot(x="total_bill", y="tip", hue="sex",style="smoker", data=tipsDF)
```

```python
sns.distplot(titanicDF[titanicDF['Survived']==0]['Age'], hist=False)
sns.distplot(titanicDF[titanicDF['Survived']==1]['Age'], hist=False)
```

```python
cross = pd.crosstab(titanicDF['Pclass'], titanicDF['Survived'])
sns.heatmap(cross)
```

```python
sns.pairplot(irisDF, hue='species')
```
