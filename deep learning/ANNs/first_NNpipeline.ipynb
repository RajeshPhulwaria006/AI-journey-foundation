```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
X = df.drop(['diagnosis'], axis=1)
y = df['diagnosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15)
```

```python
sd = StandardScaler()
X_train = sd.fit_transform(X_train)
X_test = sd.transform(X_test)
```

```python
lb = LabelEncoder()
y_train = lb.fit_transform(y_train)
y_test = lb.transform(y_test)
```

# Numpy to Tensor transformation.

```python
import torch

X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

# Neural Network defination.

```python
class myNN:
  def __init__(self, X):
    self.weights = torch.rand(X.shape[1], 1, dtype=torch.float64, requires_grad=True)
    self.bias = torch.zeros(1, dtype=torch.float64, requires_grad=True)

  def forward(self, X):
    z = torch.matmul(X, self.weights) + self.bias
    y_pred = torch.sigmoid(z)
    return y_pred

  # function to calculate loss (binary cross-entropy loss)
  def loss(self, prediction, target):
    epsilon = 1e-7  # to prevent log(0)
    y_pred = torch.clamp(prediction, epsilon, 1 - epsilon)
    loss = -(target * torch.log(y_pred) + (1 - target) * torch.log(1 - y_pred)).mean()
    return loss
```

## training pipeline.

```python
learningRate = 0.01
epochs = 500
```

```python
md = myNN(X_train_tensor)

for i in range(epochs):
  # forward pass
  y_pred = md.forward(X_train_tensor)

  # loss calculation
  loss = md.loss(y_pred, y_train_tensor)

  # backward pass
  loss.backward()

  # parameter update
  with torch.no_grad():
    md.weights -= learningRate * md.weights.grad
    md.bias -= learningRate * md.bias.grad

  # zero gradients
  md.weights.grad = None
  md.bias.grad = None

  if i % 100 == 0:
    print(f'epchs = {i}, loss = {loss.item()}')
```

```python
loss = md.loss(md.forward(X_train_tensor), y_train_tensor)

print(loss.requires_grad)
print(loss.grad_fn)
```

```python
# model evaluation
with torch.no_grad():
  y_pred = md.forward(X_test_tensor)
  y_pred = np.round(y_pred.numpy()).flatten()

accuracy = (y_pred == y_test_tensor).float().mean()
print(accuracy)
```

```python
import torch.nn as nn
```

```python
class basic_NN(nn.Module):
  def __init__(self, num_features):
    # define neual network architecture
    super().__init__()    # invoke parent class's constructor
    self.linear = nn.Linear(num_features, 1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):
    output = self.linear(features)
    output = self.sigmoid(output)
    return output
```

```python
# Built in binary cross entropy loss function.
loss_func = nn.BCELoss()
```

```python
n = X_train_tensor.shape[1]
model = basic_NN(n)

# define optimizer
optimizer = torch.optim.SGD(model.parameters(), lr=learningRate)

for i in range(epochs):
  # forward pass
  y_pred = model(X_train_tensor)

  # loss calculation
  loss = loss_func(y_pred, y_train_tensor.view(-1, 1))

  optimizer.zero_grad()

  # backward pass
  loss.backward()

  # # parameter update
  # with torch.no_grad():
  #   model.linear.weight -= learningRate * model.linear.weight.grad
  #   model.linear.bias -= learningRate * model.linear.bias.grad

  # # zero gradients
  # model.linear.weight.grad = None
  # model.linear.bias.grad = None
  optimizer.step()

  if i % 100 == 0:
    print(f'epchs = {i}, loss = {loss.item()}')
```

```python
# model evaluation
with torch.no_grad():
  y_pred = model(X_test_tensor)
  y_pred = np.round(y_pred.numpy()).flatten()

accuracy = (y_pred == y_test_tensor).float().mean()
print(accuracy.item())
```
