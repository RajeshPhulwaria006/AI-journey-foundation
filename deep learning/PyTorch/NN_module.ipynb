# Basic neural network with 5 weights and 1 bias.

```python
import torch
import torch.nn as nn

class myNN(nn.Module):
  def __init__(self, num_features):
    # define neual network architecture
    super().__init__()    # invoke parent class's constructor
    self.linear = nn.Linear(num_features, 1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):
    output = self.linear(features)
    return self.sigmoid(output)
```

```python
features = torch.rand(10, 5)
model = myNN(5)
model(features)
```

# Neural network with hidden layers.

```python
class Neural_Network(nn.Module):
  def __init__(self, num_features):
    # define neual network architecture
    super().__init__()    # invoke parent class's constructor
    self.linear1 = nn.Linear(num_features, 3)
    self.relu = nn.ReLU()
    self.linear2 = nn.Linear(3, 1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):
    output = self.linear1(features)
    output = self.relu(output)
    output = self.linear2(output)
    output = self.sigmoid(output)

    return output;
```

```python
features = torch.rand(10, 5)

md = Neural_Network(5)
md(features)
```

```python
md.linear1.weight
```

```python
md.linear2.weight
```

```python
from torchinfo import summary

summary(md, input_size=(10, 5))
```

# Sequantial NN module.

```python
class Sq_NN(nn.Module):
  def __init__(self, num_features):
    super().__init__()
    self.network = nn.Sequential(
        nn.Linear(num_features, 3),
        nn.ReLU(),
        nn.Linear(3, 1),
        nn.Sigmoid()
    )

  def forward(self, features):
    output = self.network(features)
    return output;
```

```python
newMD = Sq_NN(5)
newMD(features)
```
