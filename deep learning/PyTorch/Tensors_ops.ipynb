```python
import torch
print(torch.__version__)
```

```python
torch.cuda.is_available()
torch.cuda.device_count(), torch.cuda.get_device_name(0)
```

```python
# tensors creation
tensor = torch.empty(2, 3) # creates empty matrics with memory allocation of given size of matrcs
type(tensor)
```

```python
zeroMat = torch.zeros(2, 3) # will create matrics of initialized value - 0
onesMat = torch.ones(2, 3) # will create matrics of initialized value - 1
randMat = torch.rand(2, 3) # will create matrics of initialized value - random
print(zeroMat)
print(onesMat)
print(randMat)
```

```python
# user-defined tensor
tensor = torch.tensor([[1, 2, 3], [4, 5, 6]])
print(tensor)
```

```python
# more ways
print(torch.arange(1, 10, 1))   # create tensor in range of 1 - 10 with 1 of jumps
print(torch.linspace(1, 10, 5))  # create tensor in range of 1 - 10 with 5 elements
print(torch.logspace(1, 10, 5))
print(torch.eye(5))   # create matrics of 5x5 size
```

```python
# if i wanna create a same size tensor i had with filled value of zero or ones.
print(torch.empty_like(tensor))
print(torch.zeros_like(tensor))
print(torch.ones_like(tensor))
print(torch.rand_like(tensor.to(torch.float64)))
```

```python
# assign dtypes with tesnor creation
myTensor = torch.tensor([[1, 2, 3], [4, 5, 6]], dtype=torch.float64)
print(myTensor)
```

```python
newTensor = torch.tensor([[3.5, 2.3, 1.8], [6.6, 5.4, 4.1]], dtype=torch.float64)
print(newTensor)
```

# Scaler operations.

```python
print(myTensor + 2)
print(newTensor - 2)
print(myTensor * 2)
print(newTensor / 2)
print(myTensor ** 2)
```

# Two tensor operations.

```python
print(myTensor + newTensor)
print(myTensor - newTensor)
print(myTensor * newTensor)  # element wise
print(myTensor / newTensor)
print(myTensor ** newTensor)
```

```python
print(torch.sum(myTensor))
print(torch.sum(newTensor, dim=0))  # column wise addition
print(torch.sum(newTensor, dim=1))  # row wise addition
```

```python
print(torch.mean(newTensor))
print(torch.median(newTensor))
```

```python
print(torch.std(newTensor))
print(torch.var(newTensor))
```

```python
print(torch.min(newTensor))
print(torch.max(newTensor))
print(torch.argmin(newTensor))
print(torch.argmax(newTensor))
```

```python
# dot product / matrix multiplication
vec1 = torch.tensor([1, 2])
vec2 = torch.tensor([3, 4])
print(torch.dot(vec1, vec2))

ten1 = torch.rand(2, 3, dtype=torch.float64)
ten2 = torch.rand(3, 2, dtype=torch.float64)
print(torch.matmul(ten1, ten2))
```

```python
# cloning of tensors
a = torch.rand(2, 3)
```

```python
b = a   # initialized same value of tensor as well as memory
b
```

```python
id(a), id(b)
```

```python
b = a.clone()   # creates copy of same value of tensor as well as new memory allocation
```

```python
id(a), id(b)
```
