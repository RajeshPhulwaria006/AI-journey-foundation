# Calculating gradient of loss functions.

```python
# manual way

x = torch.tensor(6.7)   # input feature
y = torch.tensor(0.0)   # True label (target /  binary (0/1))

w = torch.tensor(1.0)   # weight
b = torch.tensor(0.0)   # bias
```

```python
# function to get loss (binary cross-entropy loss)
def loss(prediction, target):
  epsilon = 1e-8  # to prevent log(0)
  prediction = torch.clamp(prediction, epsilon, 1 - epsilon)
  return - (target * torch.log(prediction) + (1 - target) * torch.log(1 - prediction));
```

```python
# Forward pass
z = w * x + b
y_pred = torch.sigmoid(z)
bin_loss = loss(y_pred, y)
```

```python
bin_loss
```

```python
# Derivatives / Backward pass / Backpropagation
dl_dypred = (y_pred - y) / (y_pred * (1 - y_pred))
dypred_dz = y_pred * (1 - y_pred)
dz_dw = x
dz_db = 1

dl_dw = dl_dypred * dypred_dz * dz_dw
dl_db = dl_dypred * dypred_dz * dz_db
```

```python
print("Manual dw: ", dl_dw)
print("Manual db: ", dl_db)
```

# Mentos zindagi.

```python
x = torch.tensor(6.7)
y = torch.tensor(0.0)

w = torch.tensor(1.0, requires_grad=True)
b = torch.tensor(0.0, requires_grad=True)

z = w * x + b
y_pred = torch.sigmoid(z)
bin_loss = loss(y_pred, y)
```

```python
z, y_pred, bin_loss
```

```python
# Backward pass
bin_loss.backward()
```

```python
print("AutoGrad dw: ", w.grad)
print("AutoGrad db: ", b.grad)
```
