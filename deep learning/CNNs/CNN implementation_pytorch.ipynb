```python
import torch
import torch.nn as nn

# Input: (batch_size, channels, height, width)
x = torch.randn(1, 1, 5, 5)  # 5x5 image
```

```python
x
```

```python
# Case 1: No padding, stride = 1
conv1 = nn.Conv2d(in_channels=1, out_channels=1, kernel_size=3, stride=1, padding=0)
out1 = conv1(x)

print("Output shape (no padding, stride=1):", out1.shape)
# Expected: (1, 1, 3, 3)
```
