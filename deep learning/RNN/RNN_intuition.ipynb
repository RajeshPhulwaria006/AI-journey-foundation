```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer(oov_token='nothing')
```

```python
docs = [
    "I love machine learning",
    "The cat is sleeping on the sofa",
    "She drinks coffee every morning",
    "He went to the market yesterday",
    "Birds are flying in the blue sky",
    "The weather is pleasant today",
    "Raj is studying deep learning at college",
    "Artificial intelligence is changing the world",
    "The movie was interesting but the ending was unexpected",
    "If it rains tomorrow we will stay home and watch movies"
]
```

```python
tokenizer.fit_on_texts(docs)
```

```python
tokenizer.word_index
```

```python
tokenizer.word_counts
```

```python
sequences = tokenizer.texts_to_sequences(docs)
print(sequences)
```

```python
from tensorflow.keras.utils import pad_sequences
seq_tf = pad_sequences(sequences, padding='post')
```

```python
print(seq_tf)
print(seq_tf.shape)
```

## Using pytorch - manually

```python
# Step 1: Build vocabulary
vocab = {"nothing": 0}

for sentence in docs:
    for word in sentence.lower().split():
        if word not in vocab:
            vocab[word] = len(vocab)

print(vocab)
```

```python
# step 2: convert sentence into sequence
sequences = []

for sentence in docs:
    seq = [vocab[word] for word in sentence.lower().split()]
    sequences.append(seq)

print(sequences)
```

```python
import torch
from torch.nn.utils.rnn import pad_sequence

tensor_sequences = [torch.tensor(seq) for seq in sequences]

padded = pad_sequence(
    tensor_sequences,
    batch_first=True,
    padding_value=vocab["nothing"]
)

print(padded)
print(padded.shape)
```
